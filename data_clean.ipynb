##Định dạng cột num_sold

```python
def convert_to_int(value):
    if pd.isna(value):
        return 0
    value = value.replace('Đã bán','')
    # Xử lý trường hợp có chữ "K" (nghìn)
    if 'K' in value:
        value = float(value.replace('K', '')) * 1000
    return int(value)
```

```python
df['num_sold']=df['num_sold'].apply(convert_to_int)
df['num_sold']
```
