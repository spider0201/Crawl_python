```python
itemID = []
title = []
image_url =[]
price = []
ori_price = []
rating_score = []
num_reviews = []
location = []
sellerName = []
sellerID = []
brand_name = []
brand_id = []
num_sold = []
```

```python
for i in range (1,102):
    params['page'] = i
    response = requests.get('https://www.lazada.vn/ao-thun-cho-nam/', params=params, cookies=cookies, headers=headers)
    if response.status_code == 200  :
        print(f'Success page {i}!!!')
        result_json = response.json()['mods']['listItems']
        for record in result_json:
            itemID.append(record.get('itemId'))
            title.append(record.get('name'))
            image_url.append(record.get('image'))
            price.append(record.get('price'))
            ori_price.append(record.get('originalPrice'))
            rating_score.append(record.get('ratingScore'))
            num_reviews.append(record.get('review'))
            location.append(record.get('location'))
            sellerID.append(record.get('sellerId'))
            sellerName.append(record.get('sellerName'))
            brand_id.append(record.get('brandId'))
            brand_name.append(record.get('brandName'))
            num_sold.append(record.get('itemSoldCntShow'))
    else:
        print(f"Failed to fetch data for page {i}")
```

```python
df = pd.DataFrame({'ID':itemID,
                  "Title":title,
                  'url_ima':image_url,
                  'Price':price,
                'ori_price':ori_price,
                'rating':rating_score,
                'num_reviews':num_reviews,
                'Country':location,
                'sellerID':sellerID,
                'sellerName':sellerName,
               'brand_id':brand_id,
                'brand_name':brand_name,
                'num_sold':num_sold
                  })
df.head()
```
